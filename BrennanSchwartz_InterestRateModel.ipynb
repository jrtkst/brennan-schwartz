# 1. Introduction
The risk-free interest rate is the theoretical rate of return received from an investment that carried zero risk. Since the risk-free rate return can be obtained without any risk of losing the investment value, any other variety of investment should offer a higher rate of return in order to convince any investors. Therefore, the value of the risk-free rate is essential. 

To describe the risk-free rate movement, a one factor model is oftenly used. Despite the simplicity, the one factor model is known for its precision. The model is also known for being concise, compared to other models. 

The model is represented by the following stochastic differential equation
\begin{equation*}
dr(t) = a(t, r(t))\text{d}t + b(t, r(t))\;\text{d}W(t)
\end{equation*}
with the $b(t, r(t))\;\text{d}W(t)$ term as the "single factor", $W(t)$ a Brownian motion under a measure $P$, $a$ and $b$ are adapted stochastic processess, and $r(t)$ the risk-free rate at time $t$.  

Unlike investment instruments like cash, stocks, or bonds, there are 3 basic characteristics which are desirable from the risk-free rate models. 
* Risk-free rate should always be positive because if the risk-free rate goes under zero, it means that the investors would have not gain any return from their investment.
* $r(t)$ should be autoregressive to ensure that $r(t)$ cannot drift out to 0 or $\pm \infty$, but the $r(t)$ value must be kept around a long-term target as the time progresses.
* A simple formula for the model is desirable. This will simplifies the computation process. 




# 2. Brennan-Schwartz (BS) Model

Brennan and Schwartz introduced their model back in 1979 as
\begin{equation*}
dr(t) = \alpha (\mu - r(t))\text{d}t + \sigma r(t) \text{d}W(t)
\end{equation*}
The term $\alpha (\mu-r(t))$ is the drift factor. The drift form ensure that autoregressive property is achieved and $r(t)$ value will be kept around the long term mean $\mu$. The volatility term $\sigma$ is multiplied by $r(t)$. This implies as the $r(t)$ getting smaller, the randomness is reduced. This implies that "randomness" increases linearly with r(t).




> **Ito-Doeblin formula for an Ito Process**\
>Let $X(t), t\geq 0$, be an Ito process written
> $$
> X(t) = X(0) + \int_0^t \Delta (u)\text{d}W(u) + \int_0^t \Theta(u)\text{d}u
> $$
> with $X(0)$ nonrandom and $\Delta(u)$ and $\Theta(u)$ are adapted stochastic processes. Let $f(x)$ be a function for which the partial derivatives $f_t(t, x)$, $f_x(t,x)$ and $f_{xx}(t,x)$ are defined and continuous. Then for every $T\geq 0$, 
\begin{align*}
f(T, X(T)) &= f(0, X(0)) + \int_0^T f_t(u, X(u))\text{d}u + \int_0^T f_x(u, X(u))\Delta(u)\text{d}W(u) \\
&+ \int_0^T f_x(u, X(u))\Theta(u)\text{d}t +\dfrac{1}{2}\int_0^T f_{xx}(u, X(u))\Delta^2(u)\text{d}u 
\end{align*}
or in differential form,
\begin{align*}
\text{d}f(u, X(u)) &= f_t(u, X(u))\text{d}u + f_x(u, X(u))\Delta(u)\text{d}W(u) \\
&+ f_t(u, X(u))\Theta(u)\text{d}u + \dfrac{1}{2}f_{xx}(u, X(u))\Delta^2(u)\text{d}t
\end{align*}


First, derive the solution to $r(t)$. Define $f(t, r(t)) = e^{\alpha t}r(t)$ with
\begin{align*}
f_t(t, r) &= \alpha e^{\alpha t}r\\
f_r(t, r) &= e^{\alpha t}\\
f_{rr}(t,r) &= 0
\end{align*}

By using Ito-Doeblin formula, we have
\begin{align*}
\text{d}\left(e^{\alpha t}r(t)\right) &= \text{d}f(t, r(t)) \\
&= f_t(t, r(t))\text{d}t + f_r(t, r(t))\text{d}r(t) + \dfrac{1}{2}f_{rr}\text{d}r(t)r(t)\\
&= \alpha e^{\alpha t}r(t)\text{d}t + e^{\alpha t}\text{d}r(t)\\
&= \alpha e^{\alpha t}r(t)\text{d}t + e^{\alpha t}\left[\alpha (\mu - r(t))\text{d}t + \sigma r(t) \text{d}W(t)\right]\\
&= \alpha \mu e^{\alpha t}\text{d}t + e^{\alpha t}\sigma r(t)\text{d}W(t)
\end{align*}

Integrating both sides of the equation yields
\begin{align*}
\int_0^t \text{d}\left(e^{\alpha t}r(u)\right) &= \int_0^t \alpha \mu e^{\alpha u}\text{d}u + \int_0^t e^{\alpha u}\sigma r(u)\text{d}W(u)\\
e^{\alpha t}r(t) - e^0 r(0) &= \mu \left[e^{\alpha t}-1\right] + \sigma \int_0^t e^{\alpha u}r(u)\text{d}W(u)\\
\implies & r(t) = e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) + \sigma e^{-\alpha t}\int_0^t e^{\alpha u}r(u)\text{d}W(u)
\end{align*}



We'd have to utilize the discrete approximation of Ito integral to generate the last term of $r(t)$. 

Assume that $[0, t]$ is partitioned into $n$ subintervals $[t_0, t_1], [t_1, t_2], ..., [t_{n-1}, t_n]$. We have

\begin{align*}
r(t_k) &= e^{-\alpha t_k}r(0)  + \mu (1-e^{\alpha t_k}) + \sigma e^{-\alpha t_k}\int_0^{t_k}e^{\alpha u}r(u)\;\text{d}W(u)\\
&\approx e^{-\alpha t_k}r(0)  + \mu (1-e^{\alpha t_k}) + \sigma e^{-\alpha t_k}\sum_{j=0}^{k-1}e^{\alpha t_j}r(t_j)\left[W(t_{j+1})-W(t_j)\right]
\end{align*}
for $k=1, 2, ..n$.

## 2.1 Expectation and Variance


First, we compute the expectation for $r(t)$. Assuming the expectation is computed under a risk-neutral measure $\mathbb{P}$, we have
\begin{align*}
\mathbb{E}[r(t)] &= \mathbb{E}\left[e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) + \sigma e^{-\alpha t}\int_0^t e^{\alpha u}r(u)\text{d}W(u) \right] \\
&= e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) + \mathbb{E}\left[\sigma e^{-\alpha t}\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right]
\end{align*}

with 
\begin{align*}
\mathbb{E}\left[\sigma e^{-\alpha t}\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right] &= 
\mathbb{E}\left[\sigma e^{-\alpha t}\lim_{n\to \infty}\sum_{j=0}^{n-1}e^{\alpha t_j}r(t_j)[W(t_{j+1})-W(t_j)]\right]\\
&=\sigma e^{-\alpha t}e^{\alpha t_j}\lim_{n\to \infty}\sum_{j=0}^{n-1}\mathbb{E}\left[r(t_j)[W(t_{j+1})-W(t_j)]\right]\\
&=\sigma e^{-\alpha t}e^{\alpha t_j}\lim_{n\to \infty}\sum_{j=0}^{n-1}\mathbb{E}\left[r(t_j)\right]\mathbb{E}\left[W(t_{j+1})-W(t_j)\right]
\end{align*}
with $t_j = jt/n, j=1,..n$, and $n$ is the number of subintervals in $[0, t]$ partition. Since the expectation $\mathbb{E}\left[W(t_{j+1})-W(t_j)\right]$ is zero for all $j$, we have $\mathbb{E}\left[\sigma e^{-\alpha t}\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right]=0$. Thus, the expectation of $r(t)$ may be written as
\begin{align*}
\mathbb{E}[r(t)] = e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) 
\end{align*}


The variance of $r(t)$ may be found by the relation 
\begin{equation*}
\text{Var}(r(t))= \mathbb{E}\left[r^2(t)\right] - \mathbb{E}\left[r(t)\right]^2
\end{equation*}
whose first term is equal to 
\begin{align*}
\mathbb{E}\left[r^2(t)\right] &= \mathbb{E}\left[( e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) )^2 +
\sigma^2 e^{-2\alpha t}\left(\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right)^2 + 
2(e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}))\int_0^t e^{\alpha u}r(u)\text{d}W(u)  \right] \\
&= ( e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) )^2 +
\mathbb{E}\left[\sigma^2 e^{-2\alpha t}\left(\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right)^2 \right] +
\mathbb{E}\left[2(e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}))\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right]
\end{align*}

Using the similar method as above (expanding the Ito integral and taking its expectation), we found that the last term will be equal to zero. Thus it left us with the second term $\mathbb{E}\left[\sigma^2 e^{-2\alpha t}\left(\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right)^2 \right]$. Write $I(t) = \int_0^t\Delta(u)\text{d}W(u)$ with $\Delta(u) = e^{\alpha u}r(u)$ an adapted stochastic process. 

> **Ito Isometry**\
> Let $t$ be a positive constant and $\Delta(t), 0\leq t\leq T$ be an adapted stochastic process. Then if $I(t) =  \int_0^t\Delta(u)\text{d}W(u)$ , we have
\begin{equation*}
\mathbb{E}\left[I^2(t)\right] = \mathbb{E}\left[\int_0^t \Delta^2(u)\;\text{d}u\right]
\end{equation*}

Therefore by Ito isometry we have
\begin{align*}
\mathbb{E}\left[\sigma^2 e^{-2\alpha t}\left(\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right)^2 \right] 
&= \sigma^2 e^{-2\alpha t}\mathbb{E}\left[\left(\int_0^t e^{\alpha u}r(u)\text{d}W(u)\right)^2 \right] \\
&= \sigma^2 e^{-2\alpha t}\mathbb{E}\left[\int_0^t e^{\alpha u}r^2(u)\text{d}u \right] \\
&= \sigma^2 e^{-2\alpha t}\int_0^t e^{\alpha u} \mathbb{E}\left[r^2(u)\right]\text{d}u  \\
&= \left(\sigma^2 e^{-2\alpha t}\right)\left(\int_0^t e^{\alpha u} \mathbb{E}\left[r^2(u)\right]\text{d}u\right)
\end{align*}

<!-- Observe that the expectation above is a multiplication of two functions of $t$. Thus, we can not differentiate both sides of the equation to isolate  -->

Finally, the variance may be written as 
\begin{align*}
\text{Var}(r(t)) &= \mathbb{E}\left[r^2(t)\right] - \mathbb{E}\left[r(t)\right]^2 \\
&=\left[e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) \right]^2 + 
\sigma^2 e^{-2\alpha t}\int_0^t e^{\alpha u} \mathbb{E}\left[r^2(u)\right]\text{d}u   - 
\left[e^{-\alpha t}r(0) + \mu(1-e^{-\alpha t}) \right]^2 \\
&=\sigma^2 e^{-2\alpha t}\int_0^t e^{\alpha u} \mathbb{E}\left[r^2(u)\right]\text{d}u 
\end{align*}







## 2.2 Simulation

In this section, several simulation will be presented to show the behavior of the risk free interest rate $r(t)$. 


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def brownian_motion(a, b, dt):
    ''' 
    Generate Brownian Motion for [a, b] with time-step dt.
    ...

    Parameters
    ----------
    a : float
        Lower bound for time domain
    b : float
        Upper bound for time domain
    dt : float
        Time step for `[a, b]`
    
    Return
    ------
    W : array
        Array of `[W(t0), W(t1), .., W(tn)]` with n : number of subintervals
    '''
    n = int((b-a)/dt) 
    W = np.zeros((1, n+1))[0]
    Z = np.random.normal(0, 1, [1, n])[0]
    for i in range(1, n+1):
        W[i] = W[i-1] + Z[i-1]*np.sqrt(dt)
    return W
    

def r_brennan_schwartz(t, dt, r0, alpha, mu, sigma):
    '''
    Generate interest rate r(t) from Brennan-Schwarz model :
        dr(t) = alpha(mu - r(t))dt + sigma*r(t)*dW(t)
    ...
    Parameters 
    ----------
    t : float
        Denotes the desired point for r(t) to be calculated
    
    dt : float
        Time step

    r0 : float
        Initial interest rate at t=0 (r(0))

    alpha : float
        Speed parameter for drift.
    
    mu : float
        Long-run average for r(t).
    
    sigma : float
        Volatility parameter.

    Returns
    -------
    r_arr : Array
        Array of [r(t0), r(t1), r(t2), ..., r(tn)]
    '''
    Nt = int(t//dt) - 1 # number of subintervals on [0, t] partition
    ts = np.array([i*dt for i in range(Nt+1)]) # [t0, t1, t2, ..., tn]
    W_arr = brownian_motion(0, t, dt)
    r_arr = np.zeros((1, Nt+1))[0]
    r_arr[0] = r0
    I_tk = 0
    for k in range(1, Nt+1):
        I_tk  += np.exp(alpha*ts[k-1])*r_arr[k-1]*(W_arr[k]-W_arr[k-1])
        r_arr[k] = np.exp(-alpha*ts[k])*r0 + mu*(1-np.exp(-alpha*ts[k])) + sigma*np.exp(-alpha*ts[k])*I_tk
    return r_arr
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.dpi'] = 200
np.random.seed(2929)
t = 20
dt = 0.05
alpha = 0.3
mu = 0.06
sigma = 0.02
Nt = int(t//dt) - 1 # number of subintervals on [0, t] partition 
t_arr = np.linspace(0, t, Nt)
ts = np.array([i*dt for i in range(Nt+1)]) # [t0, t1, t2, ..., tn]
rt_1 = r_brennan_schwartz(t, dt, 0.05, alpha, mu, sigma) # r0 < mu
rt_2 = r_brennan_schwartz(t, dt, 0.06, alpha, mu, sigma) # r0 = mu
rt_3 = r_brennan_schwartz(t, dt, 0.07, alpha, mu, sigma) # r0 = mu
plt.plot(ts, rt_1)
plt.plot(ts, rt_2)
plt.plot(ts, rt_3)
plt.plot(ts, mu*np.ones(np.shape(ts)), linestyle='dashed', color='k')
plt.grid('on')
plt.xlabel('$t$')
plt.ylabel('$r(t)$')
plt.legend(['$r_0<\mu$', '$r_0=\mu$', '$r_0>\mu$', 'Long-Run Average $\mu$'])
plt.title('Brennan-Schwartz Interest Rate $r(t)$')
plt.show()
```

The figure above is 3 example of different trajectories of $r(t)$ with different initial states $r(0)\in \{0.05, 0.06, 0.07\}$, and parameters : $t=20, \Delta t = 0.05, \alpha = 0.3, \mu=0.06, \sigma=0.02$. 
As displayed here, as time progresses the $r(t)$ value may grow above or below the long term mean $\mu$, but it will never stray too far from it.  


Next, we perform sensitivity analysis for $\alpha$ and $\sigma$ parameters. 

```python
# Different sigma
t = 20
dt = 0.05
alpha = 0.3
sigma_1 = 0.01
sigma_2 = 0.03
mu = 0.06
Nt = int(t//dt) - 1 # number of subintervals on [0, t] partition 
t_arr = np.linspace(0, t, Nt)
ts = np.array([i*dt for i in range(Nt+1)]) # [t0, t1, t2, ..., tn]
rt_4 = r_brennan_schwartz(t, dt, 0.06, alpha, mu, sigma_1) 
rt_5 = r_brennan_schwartz(t, dt, 0.06, alpha, mu, sigma_2) 

# Different alpha
dt = 0.05
alpha_1 = 0.1
alpha_2 = 0.3
sigma = 0.02
mu = 0.06
rt_6 = r_brennan_schwartz(t, dt, 0.06, alpha_1, mu, sigma) 
rt_7 = r_brennan_schwartz(t, dt, 0.06, alpha_2, mu, sigma) 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))
ax1.plot(ts, rt_4, color='blue')
ax1.plot(ts, rt_5, color='red')
ax1.plot(ts, mu*np.ones(np.shape(ts)), linestyle='dashed', color='k')
ax1.grid('on')
ax1.set_xlabel('$t$')
ax1.set_ylabel('$r(t)$')
ax1.legend([r"$\sigma=0.01$", r"$\sigma=0.03$"])
ax1.set_title(r'Sensitivity Analysis for $\sigma$')

ax2.plot(ts, rt_6, color='blue')
ax2.plot(ts, rt_7, color='red')
ax2.plot(ts, mu*np.ones(np.shape(ts)), linestyle='dashed', color='k')
ax2.grid('on')
ax2.set_xlabel('$t$')
ax2.set_ylabel('$r(t)$')
ax2.legend([r"$\alpha=0.1$", r"$\alpha=0.3$"])
ax2.set_title(r'Sensitivity Analysis for $\alpha$')

plt.tight_layout()
plt.show()
```

We can see from the figures above that the effect of change in $\sigma$ is very distinguishable. Bigger value of $\sigma$ will produce more "aggresive" randomness on each timestep. 

For the sensitivity analysis of $\alpha$ which represents the speed of drift. Bigger $\alpha$ represents faster drift, thus the trajectory with bigger $\alpha$ looked like almost as if it was "squeezed" from the trajectory with smaller $\alpha$. 

Next we simulated 5000 different trajectories of $r(t)$ and gather the histogram for it at $t=20$ s using $r(0)=0.07$ and same parameters as above. For comparison, a normal curve with mean and standard deviation acquired from the simulation will also be displayed.

## 3. Pricing for Zero Coupon Bonds (ZCB)

**Bond** is a security that in a predetermined time (maturity time) pays a predetermined amount of money. Zero Coupon Bond (ZCB) is a type of bond that does not pay interest during the life of the bonds. 

Denote $P(t, r, T)$ as the bond price at time $t$ with maturity time $T$ with risk free interest rate $r$. Without loss of generality assume that the ZCB will pay an amount of 1 at time $t=T$, thus we have the terminal condition of $P(T, r, T) = 1$. 

Define the discount process $D(t)$ as 
\begin{equation*}
D(t) = \exp\left(-\int_0^t r(s)\;\text{d}s\right)
\end{equation*}

To obtain a relation between discount process $D(t)$ and $r(t)$, we must first define $I(t)=\int_0^t r(s)\text{d}s$ so that $\text{d}I(t)=r(t) \text{d}t$ and $\text{d}I(t) \text{d}I(t)=0$. Define $f(x)=e^{-x}$ whose derivatives are $f'(x)=-f(x), f''(x)=f(x)$. Using Ito-Doeblin formula, we have
\begin{align*}
\text{d}D(t) &= \text{d}f(I(t))\\
&= f'(I(t))\text{d}I(t) + \dfrac{1}{2}f''(I(t))\text{d}I(t)\text{d}I(t)\\
&= -f(I(t))r(t)\text{d}t\\
&= -r(t)D(t)\;\text{d}t
\end{align*}

Therefore, $\text{d}D(t) = -r(t)D(t)\text{d}t$. 

The risk-neutral pricing formula stated that the discounted price of this bond should be a martingale under the risk-neutral measure. In other words, the bond price $P(t, r, T)$ for $0\leq t\leq T$ should satisfy
\begin{align*}
D(t)P(t, r, T) &= \mathbb{E}\left[D(T)P(T, r, T) \mid \mathcal{F}(t)\right] = \mathbb{E}\left[D(T)\mid \mathcal{F}(t)\right] \\
\implies P(t, r, T)&=\mathbb{E}\left[\dfrac{D(T)}{D(t)}\mid \mathcal{F}(t)\right]\\
&=\mathbb{E}\left[\exp\left(-\int_t^T r(s)\text{d}s\right) \bigg| \mathcal{F}(t)\right]
\end{align*}

Find the differential of martingale $D(t)P(t, r, T)$ as follows
\begin{align*}
\text{d}(D(t)P(t, r, T)) &= P(t, r, T)\text{d}D(t) + D(t)\text{d}P(t, r, T)\\
&=-P(t, r, T)D(t) + D(t) \left[P_t \text{d}t + P_r \text{d}r + \dfrac{1}{2}P_{rr}\text{d}r\text{d}r\right]\\
&=D(t)\left[-Pr + P_t + \alpha(\mu-r)P_r + \dfrac{1}{2}P_{rr}\sigma^2r^2 \right]\text{d}t + D\sigma r P_r \text{d}W
\end{align*}

From the property of martingale, the $\text{d}t$ term must be equal to zero. We obtain a partial differential equation
\begin{align*}
P_t + \alpha(\mu-r)P_r + \dfrac{1}{2}P_{rr}\sigma^2r^2 = Pr
\end{align*}


## 3.1 Numerical Simulation for Bond Price $P(t, r, T)$

We will approximate $P(t, r, T)$ numerically using Crank-Nicolson method. Firstly, do a change variable from $t$ to $\tau=T-t$ in order to impose an initial condition rather than a terminal condition. We have
\begin{equation*}
\dfrac{\partial P}{\partial t}=-\dfrac{\partial P}{\partial \tau}
\end{equation*}

Now we are left with a PDE
\begin{equation*}
\dfrac{\partial P}{\partial \tau}-\alpha(\mu-r)\dfrac{\partial P}{\partial r} - \dfrac{1}{2}\sigma^2 r^2 \dfrac{\partial^2 P}{\partial r^2}=-Pr
\end{equation*}
with initial condition
\begin{equation*}
P(0, r, T) = 1
\end{equation*}
To solve the PDE numerically, we need two boundary conditions for each ends of $r$ interval. For $r=0$ (zero interest rate) it implies that the bond price should be the same as payoff in the maturity time. We write
\begin{equation*}
P(\tau, 0, T) = 1
\end{equation*}
For $r=R\in\mathbb{R}^+$, we may use the relation above as follows:
\begin{align*}
P(t, R, T)&=\mathbb{E}\left[\exp\left(-\int_t^T R\;\text{d}s\right) \bigg| \mathcal{F}(t)\right]\\
&=e^{-R(T-\tau)}
\end{align*}


To discretize the PDE above, assume a rectangular grid for $\tau$ and $r$ given by $\{i\Delta r+r_0, n\Delta \tau\}_{i=0}^{N_r},_{n=0}^{N_\tau}$ with
\begin{align*}
r_i &= r_0 + i\Delta r, \hspace{0.25cm} i=0, 1, ..., N_r\\
\tau_n &= n\Delta \tau, \hspace{0.25cm} n=0, 1, ..., N_t
\end{align*}
and $r_0=0, r_{N_r}=R, \tau_0=0, \tau_{N_\tau}=T$. 

Applying Crank-Nicolson scheme to the PDE yields
\begin{align*}
\dfrac{P_i^{n+1}-P_i^n}{\Delta \tau}  -\dfrac{\alpha(\mu-r_i)}{2}\left[\dfrac{P_{i+1}^n-P_{i-1}^n}{2\Delta r}+\dfrac{P_{i+1}^{n+1}-P_{i-1}^{n+1}}{2\Delta r}\right] -\dfrac{\sigma^2r_i^2}{4}\left[\dfrac{P_{i+1}^n-2P_i^n+P_{i-1}^n}{\Delta r^2}+\dfrac{P_{i+1}^{n+1}-2P_i^{n+1}+P_{i-1}^{n+1}}{\Delta r^2}\right] = -\dfrac{r_iP_i^n + r_iP_{i}^{n+1}}{2}
\end{align*}
with $P_i^n = P(\tau_n, r_i, T)$. 

This scheme may be rewritten in the matrix form as
\begin{align*}
\textbf{P}^{n+1} = \textbf{M}_2^{-1}\left[\textbf{M}_1\textbf{P}^n + \textbf{W}^n\right]
\end{align*}
with $\textbf{P}^n = \begin{bmatrix}P_1^n\\P_2^n\\\vdots\\P_{N_r-1}^n\end{bmatrix}\in \mathbb{R}^{N_r-1}$ and matrices
\begin{align*}
\textbf{M}_1 &= 
\begin{bmatrix}
B_1   & C_1 & 0   &  0  &  ...  &  0  &  0\\
A_2   & B_2 & C_2 &  0  &  ...  &  0  &  0\\
0     & A_3 & B_3 & C_3 &  ...  &  0  &  0\\
\vdots &     &     &     &  \ddots   &     & \vdots\\
0      &  0  &      & ... & A_{N_r-2} & B_{N_r-2} & C_{N_r-2}\\
0      &  0  &      & ... &  0        & A_{N_r-1} & B_{N_r-1}
\end{bmatrix},\\
\textbf{M}_2 &= 
\begin{bmatrix}
D_1   & -C_1 & 0   &  0  &  ...  &  0  &  0\\
-A_2   & D_2 & -C_2 &  0  &  ...  &  0  &  0\\
0     & -A_3 & D_3 & -C_3 &  ...  &  0  &  0\\
\vdots &     &     &     &  \ddots   &     & \vdots\\
0      &  0  &      & ... & -A_{N_r-2} & D_{N_r-2} & -C_{N_r-2}\\
0      &  0  &      & ... &  0        & -A_{N_r-1} & D_{N_r-1}
\end{bmatrix},\\
\textbf{W}^n &= 
\begin{bmatrix}
2A_1\\
0\\
\vdots\\
0\\
C_{N_r-1}\left(P_{N_r}^n +P_{N_r}^{n+1} \right)
\end{bmatrix}
\end{align*}
and
\begin{align*}
A_i &= -\alpha(\mu-r_i)k\Delta r + \sigma^2 r_i^2 k\\
B_i &= 1 - \dfrac{\Delta \tau r_i}{2}-2\sigma^2r_i^2 k\\
C_i &= \alpha(\mu-r_i)k\Delta r + \sigma^2 r_i^2 k\\
D_i &= 1 + \dfrac{\Delta \tau r_i}{2} + 2\sigma^2r_i^2 k\\
k   &= \dfrac{\Delta\tau}{4\Delta r^2}
\end{align*}




```python
def BS_bond_price(dt, dr, T, R, alpha, mu, sigma):
    '''
    Generate bond price P(tau, r) for (r,tau) in [0, R]x[0, T] given initial condition:
        P(0, r)=1
    and boundary conditions:
        P(tau, 0)=1, P(tau, R)=exp(-R(T-tau)) 
    with tau=T-t
    ...
    Parameters 
    ----------
    R : float
        Final interest rate 

    T : float
        Simulation time
    
    dt : float
        Time step

    alpha : float
        Speed parameter for drift.
    
    mu : float
        Long-run average for r(t).
    
    sigma : float
        Volatility parameter.

    Returns
    -------
    P_arr : Array
        Array of [P]_{i,j}
    '''
    # Generate matrices
    k = dt/(4*dr**2)
    Nr = int(R//dr)-1
    Nt = int(T//dt)-1
    A_s = np.array([-alpha*(mu-i*dr)*k*dr + (sigma*i*dr)**2*k 
                    for i in range(2, Nr)])
    B_s = np.array([1 - dt*i*dr/2 - 2*(sigma*i*dr)**2*k     
                    for i in range(1, Nr)])
    C_s = np.array([alpha*(mu-i*dr)*k*dr + (sigma*i*dr)**2*k 
                    for i in range(1, Nr-1)])
    D_s = np.array([1 + dt*i*dr/2 + 2*(sigma*i*dr)**2*k 
                    for i in range(1, Nr)])
    M1 = np.diag(B_s) + np.diag(C_s, 1) + np.diag(A_s, -1)
    M2 = np.diag(D_s) + np.diag(-C_s, 1) + np.diag(-A_s, -1)
    
    P_arr = np.zeros((Nr+1, Nt+1))

    # Imposing initial and boundary conditions
    P_arr[:, 0] = 1 # P(0, r, T) = 1
    P_arr[0, :] = 1 # P(tau, 0, T) = 1
    for n in range(Nt+1):
        tau = n*dt
        P_arr[-1, n] = np.exp(-R*(tau)) # P(tau, R, T) = exp(-R*tau)
    P0 = P_arr[1:-1, 0].reshape(-1, 1)
    Wn = np.zeros((Nr-1, 1))
    Wn[0] = (-alpha*(mu-1*dr)*k*dr + (sigma*1*dr)**2*k)*2
    M2_inv = np.linalg.inv(M2)
    # Calculating Pn
    for n in range(1, Nt+1):
        Wn[-1] = -(alpha*(mu-(Nr-1)*dr)*k*dr + (sigma*(Nr-1)*dr)**2*k)*(P_arr[-1, n-1] + P_arr[-1, n])
        P1 = M2_inv@(M1@P0 + Wn)    
        P_arr[1:-1, [n]] = P1
        P0 = P1
    return P_arr

```
